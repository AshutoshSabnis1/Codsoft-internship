```python
# Now to balance the data for Legit and fraud transaction value points
# We will use sampling for creating a new dataset of normal transaction with 492 entries being selected randomly out of 284315
normal_sample=legit.sample(n=492)
```

```python
# Now merge the two datasets for fraud and legit transactions with equal number of sample points
```

```python
new=pd.concat([normal_sample,fraud],axis=0) #Axis=0 specifies row wise joining of datasets
```

```python
new
```

```python
new["Class"].value_counts()
```

```python
new.groupby("Class").mean()
```

```python
#Here we can drop the time feature and instead use a derived column using timedelta function if pandas to represent the duration
time=pd.to_timedelta(new["Time"],unit="s")
#create the derived column
new["time_hour"]=(time.dt.components.hours).astype(int)
#now drop the time column
new.drop(columns='Time',axis=1,inplace=True)
```

```python
x=new.drop("Class",axis=1)
```

```python
y=new["Class"]
```

```python
x.shape
```

```python
#Splitting the data into training and testing data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=3,stratify=y)
```

```python
#accumulating all the column names under one variable
cols=list(x.columns.values)
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
y_pred=model.predict(X_train)
pred_test=model.predict(X_test)
```

```python
#Model Evaluation
# Creating confusion matrix
from sklearn.metrics import confusion_matrix
def Plot_confusion_matrix(Y_test,Y_pred):
    cm=confusion_matrix(Y_test,pred_test)
    plt.cf()
    plt.show()
```

```python
#Accuracy on training data
acc_score=round(accuracy_score(y_pred,Y_train)*100,2)
```

```python
print("The accuracy score for training data of our model is :",acc_score)
```

```python
Y_pred=model.predict(X_test)
acc_score=round(accuracy_score(Y_pred,Y_test)*100,2)
```

```python
print("The accuracy score of our model is :",acc_score)
```
